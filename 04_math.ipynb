## Summary Step 4: Math Basics with NumPy

In this step, I calculated statistics manually to show the math behind the data.

#### What I did: 

- I calculated the "Mean" and "STD" manually using NumPy formulas.

- I scaled the "Price" column by hand using the Z-score formula.

#### Comparing and Probability:

- I calculated "Cosine Similarity" to see the difference between high and low price cars.

- I estimated the "Probability" of finding a car priced over $50,000.

#### Final result:

- Manual mathematical verification of the dataset results.

```python
import pandas as pd
import numpy as np

# Load Features Data
df = pd.read_csv("data/cleaned/cars_features.csv")

prices = df['price']

mean = np.sum(prices) / len(prices)

# STD
diff = prices - mean
std = np.sqrt(np.sum(diff**2) / len(prices))

print("--- Math Results ---")
print(f"Mean Price: {mean}")
print(f"STD Price: {std}")

# Standardize by hand (Z-Score) 
# (value - mean) / std
df['price_scaled'] = (df['price'] - mean) / std
print("\n--- Scale Check ---")
print(df['price_scaled'].head())

# sorted price cars from low to high

df_sorted = df.sort_values('price')

# low car in first line 
low_car = df_sorted[['odometer_scaled', 'car_age_scaled']].iloc[0]

# high car in last line
high_car = df_sorted[['odometer_scaled', 'car_age_scaled']].iloc[-1]

# similarity calculation
dot_product = np.sum(low_car * high_car)
norm_low = np.sqrt(np.sum(low_car**2))
norm_high = np.sqrt(np.sum(high_car**2))

similarity = dot_product / (norm_low * norm_high)
print(f"\nSimilarity between high and low cars: {similarity}")

# probability calculation
# whats probability car > 50k ?
total_cars = len(df)
expensive_cars_count = len(df[df['price'] > 50000])
probability = expensive_cars_count / total_cars

print(f"\nProbability of car > $50,000: {probability}")
```
