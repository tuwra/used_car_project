## Summary Step 1: Loading and Cleaning the Data

In this step, I cleaned the data (used cars) to prepare it for analysis.
#### What I did: 

- I selected 8 key columns: price, year, manufacturer, model, fuel type, transmission, mileage, and condition.

- I removed missing values and duplicates so they wouldn’t affect the data.

#### I adjusted the data types:

- I changed the “Year” type to an integer and the “Mileage” type to a decimal.

#### Handling outliers:

- I removed cars priced below $500 because that price is unrealistic, and I removed cars priced above 99% because those are outliers.
- 
- I also removed cars with a mileage of 0 because that is illogical.
  
#### Final result:

- A clean dataset containing 128,752 rows.

```python
import pandas as pd
import numpy as np

# Load Data
df = pd.read_csv("Data/raw/vehicles.csv")

# Show data info 
print("--- Data Info ---")

before_rows = df.shape[0]
before_cols = df.shape[1]
print(f"Total rows: {before_rows}")
print(f"Total columns: {before_cols}")

# Function to clean data
def clean_data(data):
    # Choose columns
    columns = ["price", "year", "manufacturer", "model", "fuel", "transmission", "odometer", "condition"]
    data = data[columns].copy()
    
    # Remove missing values
    data = data.dropna()
    
    # Change column type yeat to integer & column odometer to float
    data['year'] = data['year'].astype(int)
    data['odometer'] = data['odometer'].astype(float)

    
    # Remove duplicate rows
    data = data.drop_duplicates()
    
    # Clean odometer and price from illogical values
    # Remove price that is < 500 or unrealistic (99th percentile)
    upper_limit = data['price'].quantile(0.99)
    data = data[(data['price'] >= 500) & (data['price'] <= upper_limit)]
    
    # Remove odometer if it is 0
    data = data[data["odometer"] > 0]
    
    return data

df_cleaned = clean_data(df)

# Show data info after cleaning
print("--- Data Info After Cleaning ---")

after_rows = df_cleaned.shape[0]
after_cols = df_cleaned.shape[1]
print(f"Total rows: {after_rows}")
print(f"Total columns: {after_cols}")

# Check data
print("--- Checks ---")

# Check for missing values
missing_values = df_cleaned.isnull().sum().sum()
print(f"Missing Values: {missing_values}")

# Check for logical price  
price_check = (df_cleaned['price'].min() >= 500)
print(f"Price is valid (Min >= 500): {price_check}")

# Check columns after cleaning
col_check = (after_cols == 8) # to check columns count is 8 True or False 
print(f"3. Column count is 8: {col_check}")
# Check data type for column year
print(f"Year Type: {df_cleaned['year'].dtype}")

# Save cleaned data
df_cleaned.to_csv("data/cleaned/cars_cleaned.csv", index=False)
print("\nDone! Saved to cars_cleaned.csv")
```
